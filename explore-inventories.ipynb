```python
import json
from pathlib import Path

# Path to your inventory folder
inv_path = Path(r"c:\Users\Admin\Documents\PROJECTS\SIH\data\inventories")

# Inventory files
inventory_files = [
    inv_path / "1900121_meta_inventory.json",
    inv_path / "1900121_prof_inventory.json",
    inv_path / "1900121_Rtraj_inventory.json",
    inv_path / "1900121_tech_inventory.json"
]
```

```python
for f in inventory_files:
    print(f"\n📁 File: {f.name}")
    
    with open(f, "r", encoding="utf-8") as file:
        data = json.load(file)
        
        variables = data.get("variables", {})
        
        for var_name, var_info in variables.items():
            dtype = var_info.get("dtype", "unknown")
            shape = var_info.get("shape", [])
            print(f"  • {var_name:30} | dtype: {dtype:10} | shape: {shape}")
```
